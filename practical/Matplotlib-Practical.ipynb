### Pie Charts

- **matplotlib function**: `pie(sizes)`
- Explore the `explode` parameter

```python
counts = [17, 14]
plt.figure(figsize=(4, 4))
plt.pie(counts,
        colors=['blue', 'orange'],
        labels=['Category A', 'Other categories'],
        startangle=90,
        autopct='%1.2f%%'
       )
plt.show()
```

### Subplots
Purpose: Allows us to place multiple charts in a figure
**matplotlib function**: `subplot(nrows, ncols, plot_number)`

```python
plt.figure()

for i in range(1, 7):
    plt.subplot(3, 2, i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
```

```python
dist1 = np.random.normal(42, 7, 1000)
dist2 = np.random.normal(59, 3, 1000)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(dist1)
plt.title('dist1')

plt.subplot(1, 2, 2)
plt.scatter(dist2, dist1)
plt.xlabel('dist2')
plt.ylabel('dist1')
plt.title('dist1 vs. dist2')

plt.tight_layout()
```

```python
dist1 = np.random.normal(42, 7, 1000)
dist2 = np.random.normal(59, 3, 1000)

plt.figure(figsize=(10, 4))

plt.subplot(2, 2, 1)
plt.hist(dist1)
plt.title('dist1')

plt.subplot(2, 2, 2)
plt.scatter(dist2, dist1)
plt.xlabel('dist2')
plt.ylabel('dist1')
plt.title('dist1 vs. dist2')

plt.subplot(2, 2, 3)
plt.scatter(dist2, dist1)
plt.xlabel('dist2')
plt.ylabel('dist1')
plt.title('dist1 vs. dist2')

plt.tight_layout()
```
